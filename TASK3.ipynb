```python
import pandas as pd
import matplotlib.pyplot as plt

# Specifying the new file path for the Iris dataset
iris_file_path = r'C:\Users\ADMINI\Desktop\Iris.csv'

# Loading the Iris dataset
iris_data = pd.read_csv(iris_file_path)

#colume name
column_name = 'SepalLengthCm'
plt.figure(figsize=(5,5))
plt.hist(iris_data[column_name], bins=10, color='yellow', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')
plt.show()
```
